# Differential Entropy

Continuous stochastic processes are characterized by their probability density function rather than by probabilities. Hence, the sum inside the mathematical definition of entropy becomes an integral. We call the resulting entity 'differential entropy' in order to distinguish it from the classical entropy.

\begin{align}
    H({\cal X}) = - \sum_{x \in \mathbb{X}} \Pr \{ {\cal X}=x \} \cdot \log_2 \Pr \{ {\cal X}=x \}
    \qquad \Rightarrow \qquad
    h({\cal X}) = - \int_{-\infty}^{\infty} p_{\cal X}(x) \cdot \log_2 p_{\cal X}(x) \ dx \; .
\end{align}

In contast to the classical non-negative entropy, the differential entropy can take negative values. Therefore, its interpretation is difficult and not identical to that for discrete processes. However, the difference of differential entropies as required for the computation of the mutual information can be interpreted in the susual way. We obtain

\begin{align}
    I({\cal X};{\cal Y}) = h({\cal X}) - h({\cal X} \mid {\cal Y}) \geq 0 \; .
\end{align}

In the following sections, we consider the differential entropy for some exemplary densities.

## Differential Entropy for Uniform Distribution

A uniform distribution with mean $b$ has the probability density

\begin{align}
    p_{\cal X}(x) &= 
    \begin{cases}
        \frac{1}{2a} & \text{für } |x-b| < a \\
        0 & \text{sonst} \; .
    \end{cases}
\end{align}

The average power of a process with uniform distribution amounts to

\begin{align}
    S &= \int_{-\infty}^{\infty} x^2 \cdot p_{\cal X}(x) dx
       = \frac{1}{2a} \cdot \int_{-a+b}^{a+b} x^2 dx
       = \frac{1}{2a} \cdot \left. \frac{x^3}{3} \right|_{-a+b}^{a+b}
       = \frac{a^2}{3} + b^2 \; ,
\end{align}

the variance becomes $\sigma_{\cal X}^2 = S- b^2 = \frac{a^2}{3}$. 

Its differential entropy 

\begin{align}
    h({\cal X}) &= - \int_{-a+b}^{a+b} \frac{1}{2a} \cdot \log_2 \frac{1}{2a} dx
                 = \frac{1}{2a} \cdot \log_2(2a) \cdot 2a
                 = \log_2(2a) \\
                &= \log_2 \left( 2 \sqrt{3 \sigma_{\cal X}^2} \right)
                 = \frac{1}{2} \cdot \log_2 (12 \sigma_{\cal X}^2) 
\end{align}

is independent of the mean $b$, i.e. $b$ does not contain any information. As the mean contributes to the average power, communication systems mainly use zero-mean signals. In this case power and variance are identical.


## Differential Entropy of Symmetric Triangular Distribution

W have already seen that the mean has no influence on the differential entropy. Therefore, we only consider zero-mean processes in the following derivations. A symmetric triangular distribution

\begin{align}
    p_{\cal X}(x) &= 
    \begin{cases}
        \frac{1}{a} \left( 1 - \frac{|x|}{a} \right) & \text{für } |x| \leq a \\
        0 & \text{sonst} \; .
    \end{cases}
\end{align}

corresponds to the average power

\begin{align}
    S &= \sigma_{\cal X}^2
       = \int_{-\infty}^{\infty} x^2 \cdot p_{\cal X}(x) dx
       = \frac{2}{a} \cdot \int_0^a x^2 \left( 1 - \frac{x}{a} \right) dx
       = \frac{2}{a} \cdot \left( \left. \frac{x^3}{3} \right|_0^a -  \left. \frac{x^4}{4a} \right|_0^a \right)
       = \frac{a^2}{6} \; .
\end{align}

The differential entropy becomes

\begin{align}
    h({\cal X}) 
    &= - \int_{-a}^a p_{\cal X}(x) \cdot \log_2 \left( \frac{1}{a} \left( 1 - \frac{|x|}{a} \right) \right) dx \\
    &= - \log_2 \frac{1}{a} \cdot \int_{-a}^a p_{\cal X}(x) dx
       - \int_{-a}^a p_{\cal X}(x) \cdot \log_2 \left( 1 - \frac{|x|}{a} \right) dx \\  
    &= \vdots \\
    &= \frac{1}{2} \cdot \log_2(ea^2) = \frac{1}{2} \cdot \log_2(6e\sigma_{\cal X}^2) \; .
\end{align}

## Differential Entropy of Gaussian Distribution

The average power of a zero-mean normal or Gaussian distribution

\begin{align}
    p_{\cal X}(x) &= \frac{1}{\sqrt{2 \pi \sigma_{\cal X}^2}} \cdot e^{ - \frac{x^2}{2 \sigma_{\cal X}^2} }
\end{align}

equals its variance. The differential entropy becomes

\begin{align}
    h({\cal X}) 
    &= - \int_{-\infty}^{\infty} p_{\cal X}(x) \cdot \log_2 \left( \frac{1}{\sqrt{2 \pi \sigma_{\cal X}^2}}
        \cdot e^{ - \frac{x^2}{2 \sigma_{\cal X}^2} } \right) dx \\
    &= - \log_2 \frac{1}{\sqrt{2 \pi \sigma_{\cal X}^2}} \cdot \int_{-\infty}^{\infty} p_{\cal X}(x) dx
       + \frac{\log_2 e}{2 \sigma_{\cal X}^2} \cdot \int_{-\infty}^{\infty} x^2 \cdot p_{\cal X}(x) dx \\  
    &= \frac{1}{2} \cdot \log_2(2 \pi \sigma_{\cal X}^2) + \frac{1}{2} \log_2 e \\  
    &= \frac{1}{2} \cdot \log_2(2 \pi e \sigma_{\cal X}^2) \; .
\end{align}

## Differential Entropy of Laplace Distribution

The average power of a Laplace distribution

\begin{align}
    p_{\cal X}(x) &= \frac{1}{\sqrt{2 \sigma_{\cal X}^2}} \cdot e^{ - \frac{\sqrt{2} |x|}{\sigma_{\cal X}} }
\end{align}

equals its variance $\sigma_{\cal X}^2$. The differential entropy becomes

\begin{align}
    h({\cal X}) 
    &= - \int_{-\infty}^{\infty} p_{\cal X}(x) \cdot \log_2 \left( \frac{1}{\sqrt{2 \sigma_{\cal X}^2}}
        \cdot e^{ - \frac{\sqrt{2} |x|}{\sigma_{\cal X}} } \right) dx \\
    &= - \log_2 \frac{1}{\sqrt{2 \sigma_{\cal X}^2}} \cdot \int_{-\infty}^{\infty} p_{\cal X}(x) dx
       + \frac{\sqrt{2}}{\sigma_{\cal X}} \cdot \log_2 e \cdot 2 \int_0^{\infty} x \cdot p_{\cal X}(x) dx \\  
    &= \frac{1}{2} \cdot \log_2(2 \sigma_{\cal X}^2) + \log_2 e \\  
    &= \frac{1}{2} \cdot \log_2(2 e^2 \sigma_{\cal X}^2) \; .
\end{align}

## Comparison of Differential Entropies

The diagram below compares the differential entropies determined aboce. All of them depend logarithmically on their variances. They distinguish only be constant offsets from each other. For identical varianzes, the Gaussian distribution has the largest differential entropy. It can be mathematically proven that the Gaussian distribution has the largest differential entropy among all distributions with finite power. 

The uniform distribution has the smallest differential entropy among the compared distributions. For discrete processes, the uniform distribution maximizes the entropy.

```python
import numpy as np
import matplotlib.pyplot as plt

# variances
sigmaX2_log = np.linspace(-20.0,20.0,100)
sigmaX2 = 10**(sigmaX2_log/10.0)

# differential entropy for uniform distribution
h_uniform = 0.5 * np.log2(12*sigmaX2)

# differential entropy for triangular distribution
h_triangular = 0.5 * np.log2(6*np.exp(1)*sigmaX2)

# differential entropy for normal distribution
h_gaussian = 0.5 * np.log2(2*np.pi*np.exp(1)*sigmaX2)

# differential entropy for Laplace distribution
h_laplace = 0.5 * np.log2(2*np.exp(1)**2*sigmaX2)


fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
ax.plot(sigmaX2_log, h_uniform, label='uniform')
ax.plot(sigmaX2_log, h_triangular, label='triangular')
ax.plot(sigmaX2_log, h_gaussian, label='Gaussian')
ax.plot(sigmaX2_log, h_laplace, label='Laplace')
plt.legend()
plt.grid(True)
plt.xlabel('$10 \log_{10} \sigma_{\cal X}^2$')
plt.title('differential entropies')
```

The table below compares the required average power for the considered distributions to achieve identical differential entropies. The first row shows the power gains compared to a uniform distribution. The Gaussian distribution has the largest benefit which amounts to $\frac{\pi e}{6}=1.42$. The remaining rows correspondingly show the power gains of the other distributions. Gains less than 1 obviously represent losses.
