# Capacity of Fading Channels

## Ergodic Capacity for Fast Fading Channels

For fading channels, the instantaneous SNR becomes a random variable depending on the actual fading coefficient $h[k]$. Hence, the mutual information becomes a random variable itself leading to 

\begin{equation}
    C(h) = \log_2 \left( 1 + \gamma(h) \right) = \log_2 \left( 1 + |h|^2 \frac{\Es}{N_0} \right) .
\end{equation}

In information theory, slow and fast fading channels are distinguished. For fast fading channels, long code words cover many different fading states $h[k]$, Therefore, it is meaningful to average the above instantaneous capacity over all channel states. For complex Gaussian distributed channel coefficients with average variance $\sigma_{\sproc{H}}^2$, this results in the ergodic capacity

\begin{align*}
    \overline{C} 
    &= \Exp \{ C(\sproc{H}) \} = \int_0^{\infty} p_{|\sproc{H}|^2}(\xi) \cdot \log_2(1+\xi \EsNO) d\xi \\
    &= \int_0^{\infty} \frac{1}{\sigma_{\sproc{H}}^2} e^{-\xi / \sigma_{\sproc{H}}^2} \cdot \log_2(1+\xi\EsNO) d\xi \\
    &= \textstyle \log_2(e) \cdot \exp \left( \frac{1}{\sigma_{\sproc{H}}^2 \EsNO} \right) \cdot \mathrm{expint} \left( \frac{1}{\sigma_{\sproc{H}}^2 \EsNO} \right)
     < \log_2 \left( 1 + \frac{\Es}{N_0} \right) .
\end{align*}


## Outage Probability and Capacity for Slowly Fading Channels

For slowly fading channels with a coherence time fitting the length of a code word $\mathbf{c}[k]$, each word $\mathbf{c}[k]$ experiences a different channel state. Some of them have very low SNRs, others might have very large SNRs. Consequently, the error probability depends on the fading block $k$ and averaging is not meaningful. Instead, the risk of a decoding failure called outage is considered. The probability $P_{\mathrm{out}}$ represents the probability that a targeted capacity $C$ is not supported. Assuming complex Gaussian distributed channel coefficients with average variance $\sigma_{\sproc{H}}^2$, the outage probability becomes

\begin{align*}
    \Pout &= \Pr \{ \log_2(1+\gamma) < \Cout \} 
           = \Pr \{ \gamma < 2^{\Cout}-1 \} \\
          &= \int \limits_0^{2^{\Cout}-1} \frac{1}{\sigma_{\sproc{H}}^2 \EsNO} \cdot e^{-\xi / (\sigma_{\sproc{H}}^2 \EsNO)} d\xi\\
          &=  1 - \exp \left\{ \frac{1-2^{\Cout}}{\EsNO} \right\}
\end{align*}

Correspondingly, the outage capacity $\Cout$ denotes the capacity that can be reached with a  probability $\Pout$

\begin{align}
    \Cout = \log_2 \left( 1 - \EsNO \cdot \log ( 1 - \Pout ) \right)
\end{align}


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc

save_txt = True

# define signal to noise ratio
EsN0dB = np.arange(-20,40)
EsN0 = 10**(EsN0dB/10)

# define different rates R
R = np.arange(0,10,0.1)

# capacity of complex valued AWGN channel
C_awgn = np.log2(1+EsN0)

# ergodic capacity of complex valued Rayleigh fading channel
C_rayleigh = - np.log2(np.exp(1)) * np.exp(1/EsN0) * sc.expi(-1/EsN0)

# outage capacity of complex valued Rayleigh fading channel
P_out = np.array([0.001, 0.01, 0.05, 0.1, 0.2, 0.5])
C_out = np.log2(1 - np.outer(np.log(1-P_out),EsN0))
```

```python
fig1 = plt.figure(figsize=(7,4))
ax1 = fig1.add_subplot(111)
ax1.plot(EsN0dB, C_awgn, 'k--',label='AWGN')
ax1.plot(EsN0dB, C_rayleigh, 'k-',label='$\\bar{C}$')
ax1.plot(EsN0dB, C_out[0], 'b-',label='$C_{%.1f}$' %(P_out[0]*100))
ax1.plot(EsN0dB, C_out[1], 'r-',label='$C_{%d}$' %(P_out[1]*100))
ax1.plot(EsN0dB, C_out[2], 'g-',label='$C_{%d}$' %(P_out[2]*100))
ax1.plot(EsN0dB, C_out[3], 'c-',label='$C_{%d}$' %(P_out[3]*100))
ax1.plot(EsN0dB, C_out[4], 'm-',label='$C_{%d}$' %(P_out[4]*100))
ax1.plot(EsN0dB, C_out[5], 'y-',label='$C_{%d}$' %(P_out[5]*100))
plt.xlabel('$10 \\ \\log_{10}(E_s/N_0)$ in dB')
plt.grid(True)
plt.legend()
```

We observe that

* the ergodic capacity is smaller than the AWGN capacity (fading hurts)
* the stricter the outage requirement (decreasing outage probability), the smaller is the guaranteed capacity
* For outage probability zero, the capacity is also zero $\to$ reliable communication over a fading channel is impossible!


```python
# outage capacity of complex valued Rayleigh fading channel
P_out_R = 1 - np.exp(np.outer(1-2**R,1/EsN0))
P_out_C = 1 - np.exp( (1-2**C_rayleigh) / EsN0 )

fig2 = plt.figure(figsize=(7,4))
ax2 = fig2.add_subplot(111)
ax2.plot(R, P_out_R[:,10], 'b-',label='$%d$ dB' %(EsN0dB[10]))
ax2.plot(R, P_out_R[:,15], 'r-',label='$%d$ dB' %(EsN0dB[15]))
ax2.plot(R, P_out_R[:,20], 'g-',label='$%d$ dB' %(EsN0dB[20]))
ax2.plot(R, P_out_R[:,25], 'c-',label='$%d$ dB' %(EsN0dB[25]))
ax2.plot(R, P_out_R[:,30], 'm-',label='$%d$ dB' %(EsN0dB[30]))
ax2.plot(R, P_out_R[:,35], 'y-',label='$%d$ dB' %(EsN0dB[35]))
ax2.plot(C_rayleigh[10:36:5], P_out_C[10:36:5], 'k*',label='$\\bar{C}$')
plt.xlabel('$R$')
plt.ylabel('$P_{\\mathrm{out}}$')
plt.grid(True)
plt.legend()
```

## Diversity

* Multiple propagation paths might be available due to multiple receive antennas, multiple carrier frequencies or other means
* Important: copies of frame are separately received or can be separated at receiver %(spread spectrum)
* Receiver applies maximum ratio combining (MRC) of $D$ independent paths

\begin{align*}
    y_i[l] &= h_i[l] \cdot x[l] + n[l] \\
    \bar{y}[l] &= \sum_{i=1}^D h_i^*[l] \cdot y_i[l]  = x[l] \cdot \sum_{i=1}^D |h_i[l]|^2 + \sum_{i=1}^D h_i^*[l] w_i[l]
\end{align*}

* SNR after combining is chi-square distributed with $2D$ degrees of freedom

\begin{align*}
    \bar{\gamma}[l] &= \frac{\sigmaX}{\sigmaN} \cdot \sum_{i=1}^D |h_i^*[l]|^2
    \quad \text{with} \quad
    p_{\bar{\gamma}[l]}(\xi) = \frac{\xi^{D-1}}{(D-1)! \cdot (\sigmaX / \sigmaN)^{D-1}} \cdot e^{- \frac{\xi}{\sigmaX / \sigmaN}} , \quad \xi \geq 0
\end{align*}

* Diversity decreases SNR variations and improves capacity even without CSi at transmitter





### Ergodic Capacity for Fast Fading Diversity Channels

For fast fading channels, the ergodic capacity requires to average the capacity levels with respect to the SNR distribution after combining. This is now a chi-square distribution with $2D$ degrees of freedom instead of only 2 degrees of freedom.

\begin{align*}
    \overline{C} 
    &= \Exp \{ C(\vproc{H}) \} = \int_0^{\infty} p_{\bar{\gamma}[l]}(\xi) \cdot \log_2(1 + \xi) d\xi \\
    &= \int_0^{\infty} \frac{\xi^{D-1}}{(D-1)! \cdot (\EsNO)^{D-1}} \cdot e^{- \frac{\xi}{\EsNO}} \cdot \log_2(1+\xi) d\xi \\
\end{align*}

This integral will be computed empirically by sampling the channel many times and determine the corresonding capacities.

* for fast fading channels, ergodic capacity slightly increases, but effect is not too big
* For $D=10$ paths, capacity of AWGN channel is almost reached



```python
N_paths = np.arange(1,11)
N_paths = np.append(N_paths,100)
N_runs  = 100

C_diversity = np.zeros((N_runs,len(EsN0dB),len(N_paths)))

for l,d in enumerate(N_paths):
    h2 = np.mean(np.random.normal(0, np.sqrt(0.5), size=(d,N_runs))**2 + np.random.normal(0, np.sqrt(0.5), size=(d,N_runs))**2,axis=0)
    for i,snrdB in enumerate(EsN0dB):
        SNR = 10**(snrdB/10)
        C_diversity[:,i,l] = np.log2(1 + h2 * SNR)

C_diversity_avg = np.mean(C_diversity,axis=0)

fig3 = plt.figure(figsize=(7,4))
ax3 = fig3.add_subplot(111)
ax3.plot(EsN0dB, C_diversity_avg[:,0], 'b-',label='$D=%d$' %(N_paths[0]))
ax3.plot(EsN0dB, C_diversity_avg[:,1], 'g-',label='$D=%d$' %(N_paths[1]))
ax3.plot(EsN0dB, C_diversity_avg[:,-2], 'r-',label='$D=%d$' %(N_paths[-2]))
ax3.plot(EsN0dB, C_awgn, 'k--',label='AWGN')
plt.xlabel('SNR in dB')
plt.ylabel('$\\bar{C}$')
plt.grid(True)
plt.legend()
```

### Outage Capacity for Slowly Fading Diversity Channels

For slowly fading channels, outage probability and capacity are appropriate means. As in the case without diversity, the probability that the SNR after combining is below a certain threshold has to be determined.

\begin{align*}
    P_{\mathrm{out}}^{(D)}(R) &= \Pr \{ \log_2(1+\bar{\gamma}) < R \} 
           = \Pr \left\{ \gamma < 2^{R}-1 \right\} 
           = \Pr \left\{ \| \vek{h} \|^2 \EsNO < 2^{R}-1 \right\} \\
          &= \int \limits_0^{2^{R}-1} \frac{\xi^{D-1}}{(D-1)! \cdot (\EsNO/D)^D} \cdot e^{- \frac{\xi}{\EsNO/D}} d\xi \\
          &= \frac{1}{(D-1)!} \cdot \int \limits_0^{\frac{2^{R}-1}{\EsNO/D}} \xi^{D-1} \cdot e^{- \xi} d\xi
\end{align*}

Correspondingly, the outage capacity $C_{\mathrm{out}}^{(D)}$ denotes the capacity that can be reached with a  probability $P_{\mathrm{out}}^{(D)}$.

* for slowly fading channels, ergodic capacity slightly increases, but effect is not too big
* For $D=10$ paths, capacity of AWGN channel is almost reached


```python
# outage capacity of complex valued Rayleigh fading channel with diversity and MRC
N_xi = 500
R2 = np.arange(0.01,2,0.01)
P_out_diversity_R = np.zeros((R2.shape[0],len(EsN0dB),len(N_paths)))
P_out_diversity_C = np.zeros((len(EsN0dB),len(N_paths)))

for l,d in enumerate(N_paths):
    for i,snrdB in enumerate(EsN0dB):
        SNR = 10**(snrdB/10)
        for run,r in enumerate(R2):
            int_upper = (2**r-1)/(SNR/d)
            dxi = int_upper / N_xi
            xi = np.arange(0,int_upper,dxi)
            if d==1:
                P_out_diversity_R[run,i,l] = np.sum(np.exp(-xi))*dxi
            else:
                P_out_diversity_R[run,i,l] = np.sum(xi**(d-1) * np.exp(-xi))*dxi / (np.prod(np.arange(1,d).astype(float)))

        int_upper = (2**C_diversity_avg[i,l]-1)/(SNR/d)
        dxi = int_upper / N_xi
        xi = np.arange(0,int_upper,dxi)
        if d==1:
            P_out_diversity_C[i,l] = np.sum(np.exp(-xi))*dxi
        else:
            P_out_diversity_C[i,l] = np.sum(xi**(d-1) * np.exp(-xi))*dxi / (np.prod(np.arange(1,d).astype(float)))

ptrSNR = 20
fig3 = plt.figure(figsize=(7,4))
ax3 = fig3.add_subplot(111)
#ax3.plot(R, P_out_R[:,ptrSNR], 'k-',label='$D=1$')
ax3.plot(R2, P_out_diversity_R[:,ptrSNR,0], 'k-',label='$D=1$')
ax3.plot(R2, P_out_diversity_R[:,ptrSNR,1], 'b-',label='$D=%d$' %(N_paths[1]))
ax3.plot(R2, P_out_diversity_R[:,ptrSNR,2], 'g-',label='$D=%d$' %(N_paths[2]))
ax3.plot(R2, P_out_diversity_R[:,ptrSNR,4], 'r-',label='$D=%d$' %(N_paths[4]))
ax3.plot(R2, P_out_diversity_R[:,ptrSNR,9], 'c-',label='$D=%d$' %(N_paths[9]))
ax3.plot(R2, P_out_diversity_R[:,ptrSNR,-1], 'm-',label='$D=%d$' %(N_paths[-1]))

ax3.plot(C_diversity_avg[ptrSNR,[0,1,2,4,9]], P_out_diversity_C[ptrSNR,[0,1,2,4,9]], 'k*',label='$\\bar{C}$')
plt.xlim(0,2)
#plt.ylim(0,2)
plt.xlabel('$R$')
plt.ylabel('$P_{\\mathrm{out}}$')
plt.title('SNR = %d dB' %(EsN0dB[ptrSNR]))
plt.grid(True)
plt.legend()
```

## Capacity with CSI at Transmitter

If the transmitter knowns the channel, it can distribute the power over time appropriately. For slow fading channels, each frame $l$ experiences a single fading coeffcient $h[l]$. The transmitter is allowed to allocate power $P_l$ to frame $l$ fulfilling an average power constraint $\frac{1}{L} \sum_{l=1}^L P_l = P$. The following questions arise:

* How has the power to be allocated?
* How does this change the capacity?

This leads to the following optimization problem: 

\begin{equation*}
    C = \max_{P_1, \ldots, P_L} \sum_{l=1}^L \log_2 \left( 1 + \frac{|h[l]|^2 P_l}{\sigmaN} \right)
    \quad \text{s.t.} \quad
    \frac{1}{L} \sum_{l=1}^L P_l = P
\end{equation*}


The optimal power allocation is provided by waterfilling leading to

\begin{equation*}
    P_l = \left( \frac{1}{\lambda} + \frac{\sigmaN}{|h[l]|^2} \right)^+
    \quad \text{with} \quad
    ( x )^+ = \max (0, x)
\end{equation*}

The parameter $\lambda$ is determined such that the average power constraint 

\begin{equation*}
    \frac{1}{L} \sum_{l=1}^L \left( \frac{1}{\lambda} + \frac{\sigmaN}{|h[l]|^2} \right)^+ = P
end{equation*}

is fulfilled. For finite $L$, this requires the transmitter to know future channel realizations (non-causal!). For $L \to \infty$, ergodicity helps because empirical average becomes expectation

\begin{equation*}
    P(h) = \left( \frac{1}{\lambda} + \frac{\sigmaN}{|h|^2} \right)^+
    \quad \text{and}  \quad
    \Exp \left\{ \left( \frac{1}{\lambda} + \frac{\sigmaN}{|h[l]|^2} \right)^+ \right\} = P
\end{equation*}

* At high SNR, waterfilling will not provide large gains
* At low SNR, only the few best channels are used and a big gain is expected



```python
def water_filling(gain, sigma2, power=1):
    """
        Calculates the water level that touches the worst channel (the higher
        one) and therefore transmits zero power in this worst channel. After
        that, calculates the power in each channel (the vector 'Ps') for this
        water level. If the sum of all of these powers in 'Ps' is less then
        the total available power, then all we need to do is divide the
        remaining power equally among all the channels (increase the water
        level). On the other hand, if the sum of all of these powers in 'Ps'
        is greater then the total available power then we remove the worst
        channel and repeat the process. 
        Calculates minimum water-level $\\mu$ required to use all channels

        Args:
            gain (np.ndarray): _description_
            sigma2 (float): _description_

        Returns:
            _type_: _description_
    """
    P = power
    NChannels = gain.shape[0]
    RemoveChannels = 0

    ptr_sort = np.flipud(np.argsort(gain))
    gain_sort = gain[ptr_sort]

    minMu = sigma2 / (gain_sort[NChannels - RemoveChannels - 1])
    Ps = (minMu - sigma2 / (gain_sort[:NChannels - RemoveChannels]))

    while (sum(Ps) > P) and (RemoveChannels < NChannels):
        RemoveChannels += 1
        minMu = sigma2 / (gain_sort[NChannels - RemoveChannels - 1])
        Ps = minMu - sigma2 / (gain_sort[:NChannels - RemoveChannels])

    # Distributes the remaining power among the all the remaining channels
    Pdiff = P - np.sum(Ps)
    Paux = Pdiff / (NChannels - RemoveChannels) + Ps

    # Put optimum power in the original channel order
    Palloc = np.zeros(NChannels)
    Palloc[ptr_sort[:NChannels - RemoveChannels]] = Paux

    return Palloc
```

```python
#SNRdB = [-10,0,10,30]
SNRdB = EsN0dB

N_channels = [10,20,100]
N_runs     = 100

C_uni = np.zeros((N_runs,len(SNRdB),len(N_channels)))
C_wf  = np.zeros_like(C_uni)
C_zf  = np.zeros_like(C_uni)

for l,n in enumerate(N_channels):
    h2 = np.random.normal(0, np.sqrt(0.5), size=(n,N_runs))**2 + np.random.normal(0, np.sqrt(0.5), size=(n,N_runs))**2
    for i,snrdB in enumerate(SNRdB):
        SNR = 10**(snrdB/10)
        for run in range(N_runs):
            # capacity without Tx-CSI
            C_uni[run,i,l] = np.mean(np.log2(1 + h2[:,run] * SNR))
            # water filling solution
            Pwf = water_filling(h2[:,run], 1/SNR, n)
            #print(np.sum(P))
            C_wf[run,i,l] = np.mean(np.log2(1 + h2[:,run] * Pwf * SNR))
            # channel inversion
            P = 1/h2[:,run]
            P /= np.sum(P)/n
            C_zf[run,i,l] = np.mean(np.log2(1 + h2[:,run] * P * SNR))

C_uni_avg = np.mean(C_uni[:,:,2],axis=0)
C_wf_avg  = np.mean(C_wf,axis=0)
C_zf_avg  = np.mean(C_zf,axis=0)

fig3 = plt.figure(figsize=(7,4))
ax3 = fig3.add_subplot(111)
ax3.plot(SNRdB, C_uni_avg, 'k-',label='uni')
ax3.plot(SNRdB, C_awgn, 'k--',label='AWGN')
ax3.plot(SNRdB, C_wf_avg[:,0], 'b-',label='WF-%d' %(N_channels[0]))
ax3.plot(SNRdB, C_wf_avg[:,1], 'g-',label='WF-%d' %(N_channels[1]))
ax3.plot(SNRdB, C_wf_avg[:,2], 'r-',label='WF-%d' %(N_channels[2]))
ax3.plot(SNRdB, C_zf_avg[:,0], 'b--',label='ZF-%d' %(N_channels[0]))
ax3.plot(SNRdB, C_zf_avg[:,1], 'g--',label='ZF-%d' %(N_channels[1]))
ax3.plot(SNRdB, C_zf_avg[:,2], 'r--',label='ZF-%d' %(N_channels[2]))
plt.xlabel('SNR in dB')
plt.ylabel('$C$')
plt.grid(True)
#plt.ylim(0,7)
plt.legend()
```
