```python
import numpy as np
import matplotlib.pyplot as plt

# definition of Gaussian a posteriori distributions' parameters
# mean
mu = [-0.7, 1.0]
# variances
sigma2 = [0.15,0.25]
# prior probabilities
P = [0.7]
P.append(1-P[0])

dx = 0.01
x = np.arange(-4,4,dx)
g1 = np.exp(-(x-mu[0])**2/2/sigma2[0])
g2 = np.exp(-(x-mu[1])**2/2/sigma2[1])
g = P[0]*g1 + P[1]*g2

fig1 = plt.figure(figsize=(6,3))
ax1 = fig1.add_subplot(111)
ax1.plot(x, g1, 'b-',label='g1')
ax1.plot(x, g2, 'g-',label='g2')
ax1.plot(x, g, 'r-',label='g')
plt.xlabel('$x$')
plt.grid(True)
plt.legend()
```

```python
# MMSE
mmse = np.sum(x*g)*dx
mmse_ptr = np.argmin(np.abs(x-mmse))
# median
median = np.median(g)
median_ptr = np.argmin(np.abs(x-median))
# maximum a posteriori 
app_ptr = np.argmax(g) 

fig2 = plt.figure(figsize=(6,3))
ax2 = fig2.add_subplot(111)
ax2.plot(x, g, 'r-',label='g')
ax2.plot(mmse, g[mmse_ptr], 'b+',label='MMSE')
ax2.plot(median, g[median_ptr], 'gx',label='median')
ax2.plot(x[app_ptr], g[app_ptr], 'cs',label='APP')
plt.xlabel('$x$')
plt.grid(True)
plt.legend()
```
