# Multi-User Diversity

## Fading Channels with Full Channel State Information (CSI) at Transmitter(s)

Usually, channel knowledge is assumed to be available only only at the receiver by appropriate channel estimation. The CSI enables the receiver to equalize the channel for a reliable data detection. Now, we investigate a scenario in which the transmitter(s) have full CSi as well. The question to be addressed is whether this changes the system capacity and how large are the potential gains.


### Multi-User Uplink Scenario

For a multi-user uplink, the received signal at the base  and time $m$ can be described by

\begin{equation}
    y[m] = \sum_{k=1}^K h_k[m] \cdot \alpha_k[m] \cdot x_k[m] + n[m] \ .
\end{equation}

Fading channels are characterized by their coherence time, i.e. the duration for which the impulse response is almost constant. We consider $L$ of these coherence intervals and assume that the channels have independent coefficients in each period. Hence, the coefficient $h_k[m] = h_{k,l}$ of channel $k$ remains constant within coherence period $l$ (as long as time $m$ belongs to period $l$). Different $h_{k,l}$, $1 \leq l \leq L$ and $1 \leq k \leq K$, are modeled as independent random variables. The users can now allocate their powers $\alpha_{k,l}^2$ over time with an average power constraint $\frac{1}{L} \sum_{l=1}^L \alpha_{k,l}^2 = \alpha_{\max}^2$. The question to be answered is: How shall powers be allocated to maximize sum rate?

In symmetric scenario, all user channels have identical average powers, i.e.

\begin{equation} 
    \Exp \left\{ |\sproc{H}_k|^2 \right\} = \sigma_{\sproc{H}}^2 \quad \forall \; 1 \leq k \leq K
\end{equation}

holds. If we focus on the sum rate $\sum_k R_k$ and relax the individual power constraints from above to a sum power constraint

\begin{equation}
    \frac{1}{L} \cdot \sum_{l=1}^L \sum_{k=1}^K \alpha_{k,l}^2 = K \cdot \alpha_{\max}^2 \ ,
\end{equation}

the optimization problem can be written as

\begin{equation}
    C_{\mathrm{sum}} = \max_{\alpha_{k,l}^2} \frac{1}{L} \sum_{l=1}^L \log_2 \left( 1 + \frac{ \sum_{k=1}^K \alpha_{k,l}^2 |h_{k,l}|^2}{\sigmaN} \right) \; .
\end{equation}

When we look at a particular coherence period $l$, the sum rate  is maximized if the entire power is given to the user with the strongest channel. This is called opportunistic transmission. The maximum sum rate in period $l$ is

\begin{equation}
    C_{\mathrm{sum}}[l] = \log_2 \left( 1 + \frac{ \sum_{k=1}^K \alpha_{k,l}^2 |h_{k,l}|^2}{\sigmaN} \right) \; .
\end{equation}

Since we have $L$ coherence periods, the users apply waterfilling over those periods in which they have strongest channel leading to

\begin{equation}
    \alpha_{k,l}^2 = 
    \begin{cases}
        \left( \frac{1}{\lambda} - \frac{\sigmaN}{\max_{i} |h_{i,l}|^2} \right)^+ & \text{if } |h_{k,l}| = \max_i |h_{i,l}| \\
        0 & \text{else} \; .
    \end{cases}
\end{equation}

The parameter $\lambda$ has to be adjusted such that sum power constraint is fulfilled. It has to be emphasized that the waterfilling solution has to be determined w.r.t. the statistics of the strongest channel, i.e. the statistics of the strongest channel has to be determined.

For the multi-user downlink scenario, the received symbol at user $k$ and time $m$ has the form

\begin{equation}
    y_k[m] = h_k[m] \cdot \sum_{i=1}^K \alpha_i[m] \cdot x_i[m] + n_k[m] \ .
\end{equation}

The sum rate with CSI only at receiver is

\begin{equation}
    \sum_{k=1}^K R_k < \Exp_{\sproc{H}} \left\{ \log_2 \left( 1 + \frac{|h|^2 \alpha_{\max}^2}{\sigmaN} \right) \right\}
\end{equation}

As for the uplink, the sum rate is maximized by spending power only to the user with the strongest channel at a time. Hence, the same solution as for the uplink is obtained. 

It has to be noted that the multiple users do not represent a new dimension for which the optimal solution can be obtained by simple waterfilling over time and users. Instead, only the user with the strongest channel per coherence period transmits and waterfilling is applied only over the best users in $L$ coherence periods.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc

def water_filling(gain, sigma2, power=1):
    """
        Calculates the water level that touches the worst channel (the higher
        one) and therefore transmits zero power in this worst channel. After
        that, calculates the power in each channel (the vector 'Ps') for this
        water level. If the sum of all of these powers in 'Ps' is less then
        the total available power, then all we need to do is divide the
        remaining power equally among all the channels (increase the water
        level). On the other hand, if the sum of all of these powers in 'Ps'
        is greater then the total available power then we remove the worst
        channel and repeat the process. 
        Calculates minimum water-level $\mu$ required to use all channels

        Args:
            gain (np.ndarray): _description_
            sigma2 (float): _description_

        Returns:
            _type_: _description_
    """
    P = power
    NChannels = gain.shape[0]
    RemoveChannels = 0
    
    ptr_sort = np.flipud(np.argsort(gain))
    gain_sort = gain[ptr_sort]
    
    minMu = sigma2 / (gain_sort[NChannels - RemoveChannels - 1])
    Ps = (minMu - sigma2 / (gain_sort[:(NChannels - RemoveChannels)]))

    while (sum(Ps) > P) and (RemoveChannels < NChannels):
        RemoveChannels += 1
        minMu = sigma2 / (gain_sort[NChannels - RemoveChannels - 1])
        Ps = minMu - sigma2 / (gain_sort[:(NChannels - RemoveChannels)])

    # Distributes the remaining power among the all the remaining channels
    Pdiff = P - np.sum(Ps)
    Paux = Pdiff / (NChannels - RemoveChannels) + Ps

    # Put optimum power in the original channel order
    Palloc = np.zeros(NChannels)
    Palloc[ptr_sort[:(NChannels - RemoveChannels)]] = Paux

    return Palloc
```

```python
save_txt = True

# define signal to noise ratio
SNRdB    = np.arange(-10,20)

N_user = 50
N_periods = 10
N_runs = 100

C_users = np.zeros((N_user,N_periods,SNRdB.shape[0],N_runs))
C_sumTx = np.zeros((N_periods,SNRdB.shape[0],N_runs))
C_sumRx = np.zeros((N_periods,SNRdB.shape[0],N_runs))

for run in range(N_runs):
    H = np.random.normal(0, np.sqrt(0.5), size=(N_user,N_periods))**2 + np.random.normal(0, np.sqrt(0.5), size=(N_user,N_periods))**2
    
    best_user_period = np.argmax(H,axis=0)
    Hbest = np.amax(H,axis=0)
    
    for runSNR in range(SNRdB.shape[0]):
        SNR = 10**(SNRdB[runSNR]/10)
        # sum rate for CSI only at receiver
        C_sumRx[:,runSNR,run] = np.log2(1 + np.sum(H,axis=0) * SNR/N_user)
        
        # waterfilling for CSI at transmitter
        P = water_filling(Hbest, 1/SNR, N_periods)
        # user rates for CSI at transmitters
        for periods in range(N_periods):
            C_users[best_user_period[periods],periods,runSNR,run] = np.log2(1 + P[periods] * H[best_user_period[periods],periods] * SNR)
        # sum rate for CSI at transmitters
        C_sumTx[:,runSNR,run] = np.sum(C_users[:,:,runSNR,run],axis=0)

C_users_avg = np.mean(np.mean(C_users,axis=-1),axis=1)
C_sumTx_avg = np.mean(np.mean(C_sumTx,axis=-1),axis=0)
C_sumRx_avg = np.mean(np.mean(C_sumRx,axis=-1),axis=0)

C_awgn = np.log2(1+10**(SNRdB/10))
```

```python
fig1 = plt.figure(figsize=(7,4))
ax1 = fig1.add_subplot(111)
ax1.plot(SNRdB, C_users_avg.T, 'b-', linewidth = 0.1)
ax1.plot(SNRdB, C_sumTx_avg, 'r-',label='$C_{\\mathrm{sum}}$, Tx')
ax1.plot(SNRdB, C_sumRx_avg, 'g-',label='$C_{\\mathrm{sum}}$, Rx')
ax1.plot(SNRdB, C_awgn, 'k--',label='AWGN')
plt.xlabel('SNR in dB')
plt.ylabel('$C$')
plt.grid(True)
plt.legend()
```

* With CSI available only at receiver, uplink performance can reach at most AWGN capacity  
    * For $K=2$ users, gap to AWGN curve is still significant
    * For $K=20$ users, gap to AWGN curve almost vanishes
* For full CSI available at transmitter, performance of AWGN channel can be even exceeded
    * Opportunistic scheduling: serve only best user at a time
    * Share of sum rate for single user shrinks with growing $K$


Further Observations:
* Symmetric scenarios: solution is asymptotically fair because all users have the best channel with identical probability
    * for few coherence periods, large variations between user rates possible
    * asymptocially (many coherence periods or many repetitions), solution is fair
* Asymptotic sum rate with index of strongest user $k^*(\vek{h})$
\begin{equation}
    C_{\mathrm{sum}} = \Exp_{h_{k^*(\vek{h})}} \left\{  \log_2 \left( 1 + \frac{ P_{k^*(\vek{h})} |h_{k^*(\vek{h})}|^2}{\sigmaN} \right) \right\}
\end{equation}
    waterfilling to be determined w.r.t. statistics of strongest channel (order statistics)
* Asymmetric scenarios: solution might becomes unfair because users with weak channels might never get the chance to transmit



```python
SNRdB = 0
SNR = 10**(SNRdB/10)

N_user = np.arange(1,50)
#N_user = np.array([1,1])
N_periods = 50
N_runs = 1000

C2_sumTx = np.zeros((N_user.shape[0],N_periods,N_runs))

for run in range(N_runs):
    for runU,users in enumerate(N_user):
        H = np.random.normal(0, np.sqrt(0.5), size=(users,N_periods))**2 + np.random.normal(0, np.sqrt(0.5), size=(users,N_periods))**2
    
        best_user_period = np.argmax(H,axis=0)
        Hbest = np.amax(H,axis=0)
    
        # waterfilling for CSI at transmitter
        P = water_filling(Hbest, 1/SNR, N_periods)
        # user rates for CSI at transmitters
        C2_users = np.zeros((users,N_periods))
        for periods in range(N_periods):
            C2_users[best_user_period[periods],periods] = np.log2(1 + P[periods] * H[best_user_period[periods],periods] * SNR)
        # sum rate for CSI at transmitters
        C2_sumTx[runU,:,run] = np.sum(C2_users,axis=0)

C2_sumTx_avg = np.mean(np.mean(C2_sumTx,axis=-1),axis=-1)

C2_awgn = np.log2(1+SNR)
```

We observe that

* the ergodic capacity is smaller than the AWGN capacity (fading hurts)
* the stricter the outage requirement (decreasing outage probability), the smaller is the guaranteed capacity
* For outage probability zero, the capacity is also zero $\to$ reliable communication over a fading channel is impossible!


```python
fig2 = plt.figure(figsize=(7,4))
ax2 = fig2.add_subplot(111)
ax2.plot(N_user, C2_sumTx_avg, 'r-',label='$C_{\mathrm{sum}}$')
ax2.plot(N_user, C2_sumTx_avg/N_user, 'b-',label='$C_k$')
ax2.plot(N_user, C2_awgn*np.ones_like(N_user), 'k--',label='AWGN')
plt.xlabel('$K$')
plt.ylabel('$C$')
plt.grid(True)
plt.legend()
```

We observe that

* the outage probability increases with the desired rate $R$
* the ergodic capacitys for different SNRs are indicated by markers,they are not associated with $\Pout=50\%$
